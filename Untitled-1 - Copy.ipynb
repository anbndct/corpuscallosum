```python
import imageio
import matplotlib.pyplot as plt
import numpy as np
import imageio
import scipy.ndimage as ndi
from skimage import color, img_as_float, exposure, filters, morphology, io, img_as_ubyte, draw
```

```python
im = imageio.imread(r"C:\Users\50232\Downloads\mri.jpg")

print('Data type:', im.dtype)
print('Min. value:', im.min())
print('Max value:', im.max())
print('shape:', im.shape)

# Plot the grayscale image
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.colorbar();
```

```python
im_adapt = exposure.equalize_adapthist(im, clip_limit=0.03)
imguint = img_as_ubyte(im_adapt)
gray = color.rgb2gray(imguint)
img_adapteq= im @ [0.2126, 0.7152, 0.0722]

print('Type:', type(img_adapteq))
print('dtype:', img_adapteq.dtype)
print('shape:', img_adapteq.shape)
print('size:', img_adapteq.size)

plt.imshow(img_adapteq, cmap='gray')
plt.axis('off')
plt.colorbar();
```

```python
# Filter Median

Filter_Median = ndi.median_filter(img_adapteq, size=5)
fig, ax = plt.subplots(1, 1, figsize=(5, 5))  
ax.imshow(Filter_Median, cmap='gray')            
ax.set_title("Median Filter 7x7")
plt.show()
```

```python
mask_bone = Filter_Median>= 200
mask_bone
```

```python
mask_bone = mask_bone*1 
mask_bone
```

```python
im_bone = np.where(mask_bone, Filter_Median, 0)
fig, axes = plt.subplots(1, 2, figsize=(10,10))
axes[0].imshow(mask_bone, cmap='gray')
axes[1].imshow(im_bone, cmap='gray')
plt.show()
```

```python
binary_im_bone = im_bone > 0

# ROI mask
roi_mask = np.zeros_like(binary_im_bone, dtype=bool)
roi_mask[150:350, :] = True  # Adjust the coordinates to cover only the desired area

selected_area = morphology.remove_small_objects(binary_im_bone & roi_mask, min_size=500)

final_mask = np.where(roi_mask, selected_area, binary_im_bone)
final_mask = morphology.remove_small_objects(final_mask, min_size=750)
plt.figure(figsize=(10, 10))
plt.imshow(final_mask, cmap='gray')
plt.title('Final Segmented Image with Selective Blob Removal')
plt.axis('off')
plt.show()
```
